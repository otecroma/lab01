```python
import pandas as pd     # lib pandas
import numpy as np      # lib numpy

url ="https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"       #       

columns_to_replace_zero = ['times_pregnant', 'plasma_glucose', 'blood_pressure','triceps_skin', 'serum_insulin', 'body_mass_index','diabetes_pedigree_function', 'age','class_variable']       #list      

#    CSV-   URL-  DataFrame info
#       columns_to_replace_zero
info = pd.read_csv(url, names=columns_to_replace_zero)

#       DataFrame info
print("Average:", "\n", info.mean(), "\n")

#       DataFrame info
print("Variance:" , "\n", info.var(), "\n")

#        DataFrame info
print("Standard Deviation:", "\n", info.std(), "\n")


#  
for column in column_names:
    plt.figure(figsize=(9, 5))
    plt.hist(dataset[column].dropna(),  color='black')
    plt.title(f'{column}')
    plt.grid(True)
    plt.show()
```
